```python
def get_artist_uri(query):
    search = sp.search(query)
    for item in search['tracks']['items']:
        if (item['album']['artists'][0]['name']).lower() == query.lower():
            return item['album']['artists'][0]['uri']
```

```python
def get_related_artists(artist_uri):
    d = {}
    
    try:
        for item in sp.artist_related_artists(artist_uri)['artists']:
            d[item['uri']] = item['name']
        return d
    except:
        return d
```

```python
def generate_artists(artist_name, num_artists):
    artist_uri = get_artist_uri(artist_name)
    Q = deque()
    visited = []
    Q.append((artist_uri, artist_name))
    while len(Q) != 0 and len(visited)<num_artists:
        v = Q.popleft()
        if v not in visited:
            visited.append(v)
            related_artists = get_related_artists(v[0])
            Q.extend(related_artists.items())
    return visited
```

```python
def get_genres(df, column_name='artist_uri'):
    print("getting genres of each song...")
    for i, row in df.iterrows():
        print(row['artist'])
        token = spotipyxx.get_token('nwaters5')
        sp = spotipy.Spotify(token)
        try:
            for j in sp.artist(row[column_name])['genres']:
                df.at[i, j] = 1
        except:
            continue
        df.at[i, 'popularity'] = sp.artist(row[column_name])['popularity']
    return df.fillna(0)
```

```python
def get_top_tracks(artist_uri, artist):
    res = []
    try:
        top_tracks = sp.artist_top_tracks(artist_uri)
    except:
        return res
    for track in top_tracks['tracks']:
        res.append((artist_uri, artist, track['uri'], track['name']))
    
    return res
```

```python
def make_track_df(df):
    tracks = []
    for _, row in df.iterrows():
        print(_)
        tracks = tracks + get_top_tracks(row['artist_uri'], row['artist'])
    return tracks
```

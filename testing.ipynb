```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
for i, row in tre_library.iterrows():
    tre_library.at[i, 'artist - track'] = str(row['artist']) + " - " + str(row['track'])
```

```python
tre_library
```

```python
tre_matrix = tre_library.drop(columns=['artist_uri', 'track_uri', 'artist', 'track'])
```

```python
tre_matrix.set_index('artist - track', drop=True, inplace=True)
```

```python
tre_matrix.drop(columns=['key', 'mode'], inplace=True)
```

```python
tre_matrix.drop(columns=['popularity'], inplace=True)
```

```python
cosine_sim = cosine_similarity(tre_matrix, tre_matrix)
```

```python
cosine_sim
```

```python
import pandas as pd
indices = pd.Series(tre_matrix.index)

#  defining the function that takes in movie title 
# as input and returns the top 10 recommended movies
def recommendations(title, cosine_sim = cosine_sim):
    
    # initializing the empty list of recommended movies
    recommended_songs = {}
    
    # gettin the index of the movie that matches the title
    idx = indices[indices == title].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    # getting the indexes of the 10 most similar movies
    top_10_indexes = list(score_series.iloc[1:51].index)
    top_10_scores = list(score_series.iloc[1:51])
    # populating the list with the titles of the best 10 matching movies
    for i, j in zip(top_10_indexes, top_10_scores):
        recommended_songs.update({list(tre_matrix.index)[i]: j})
        
        
    return recommended_songs
```

```python
recommendations("Cuco - Drown")
```
